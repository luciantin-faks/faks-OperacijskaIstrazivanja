* Konstante
   * cijena proizvodnje (odredeno u tablici)
   * provizija (npr. 10%)
   * cijena rada (odredeno u tablici)
   * broj uzvanika
   * budžet

<br>


* Varijable
   * \[ Stavka 1 - Stavka n \] - kolicina stavke n
   
<br>
 
* Funkcija cilja
    * Max( ukupna cijena sa provizijom - ukupna cijena) 
        * maximiziramo profit 

    * Max( <br>
        suma( provizija *  (cijena sastojaka * Stavka n + cijena rada * Stavka n) ) // od 1 do n  <br>
        \- suma((cijena sastojaka * Stavka n + cijena rada * Stavka n) ) <br>
        ) 

<br>

* Uvjeti
    * suma kolicine hrane/pica >= broj uzvanika (posebno za hranu i pice)
    * hrana n / suma hrane  <= 0.333..   // udio mora biti manji od 1/3 
        * udio odredenog jela nesmije biti veci od 33.3 % (barem 3 vrste jela), ako prijede onda ce netko imati manje vrsta jela
    * pice n / suma pica  <= 0.5   // udio mora biti manji od 1/2, isto kao za hranu
    * suma( provizija *  (cijena sastojaka * Stavka n + cijena rada * Stavka n) )  <=  budžet
    
    
<br>
