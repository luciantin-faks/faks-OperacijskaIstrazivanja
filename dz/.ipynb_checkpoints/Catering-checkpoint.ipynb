### Konstante
   * cijena proizvodnje (odredeno u tablici)
   * provizija (npr. 20%, na cijenu stavke)
   * cijena rada (odredeno u tablici)
   * broj uzvanika
   * budžet

<br>


### Varijable
   * Stavka\[x\] - kolicina stavke x
   
<br>
 
### Funkcija cilja
   * Max( ukupna cijena sa provizijom - ukupna cijena) 
        * maximiziramo profit 

   * Max( suma( provizija * ukupna_cijena_stavke[x] - ukupna_cijena_stavke[x] ))

<br>


    
<br>

### Uvjeti 

* kolicina hrane/pica
    * donja granica  <br><br>
        * ukupna_kolicina_pica   >= broj_uzvanika * min_kolicina_pica_po_uzvaniku
        * ukupna_kolicina_hrane  >= broj_uzvanika * min_kolicina_hrane_po_uzvaniku <br><br>
    * gornja granica  (donja + 5%) , mozemo napraviti max 5% vise hrane/pica, ako se makne onda dode tocno 300 ukupno, ali to je da kao bude mrvicu vise za svaki slucaj <br><br>
        * ukupna_kolicina_pica   <= broj_uzvanika * min_kolicina_pica_po_uzvaniku  * 1.05
        * ukupna_kolicina_hrane  <= broj_uzvanika * min_kolicina_hrane_po_uzvaniku * 1.05

<br>

* udio odredenog jela nesmije biti veci od 33.3 % (barem 3 vrste jela), jer ako prijede onda ce netko imati manje vrsta jela, isto i za pice samo 50%

* min udio tako da ima vise izbora



* max udio kolicine hrane 
    * kolicina_stavke[x] <= ukupna_kolicina_hrane * max_udio_kolicine_hrane (1/3)


* min udio kolicine hrane 
    * kolicina_stavke[x] <= ukupna_kolicina_hrane * min_udio_kolicine_hrane (0.025)



* max udio kolicine pica 
    * kolicina_stavke[x] <= ukupna_kolicina_pica * max_udio_kolicine_pica (1/2)


* min udio kolicine pica 
    * kolicina_stavke[x] <= ukupna_kolicina_pica * min_udio_kolicine_pica (0.025)


<br>

* Ukupna cijena (provizija * cijena (za svaku stavku ukupno, ne ukupno za sve) ) mora biti manja od troska
    *  provizija * ukupna_cijena_stavke[x] <= max_trosak

```python
from pymprog import * 
import matplotlib.pyplot as plt
import numpy as np
import math

begin('basic')


######################################################################################
# Podaci

cijena_minute = 35/60
broj_uzvanika = 100
provizija = 1.2 # trosak * provizija   (provizija 20 %)
max_trosak = 5000
broj_kom_mnozitelj = 5 # radi se po 5 kom/porcija odjednom

min_kolicina_pica_po_uzvaniku  = 2
min_kolicina_hrane_po_uzvaniku = 3


max_udio_kolicine_pica = 1/min_kolicina_pica_po_uzvaniku
max_udio_kolicine_hrane = 1/min_kolicina_hrane_po_uzvaniku

min_udio_kolicine_pica  = 0.025
min_udio_kolicine_hrane = 0.025


cijena_sastojaka_ = {
    "sok od rajcice" : 12,
    "vodka" : 40,
    "gin" : 45,
    "tonik" : 12,
    "vermouth" : 35,
    "kruh": 16,
    "sir": 45,
    "mortadela": 65,
    "vrecica juhe": 90,
    "pasta": 12,
    "maslac": 80,
    "rajcica konzerva": 8,
    "cevapi": 28,
    "brasno": 6,
}

kol_sastojaka = {
    "sok od rajcice":   {"Bloody Mary":0.09 * 5,"Gin & Tonic":0,"Martini":0,"Bloody gin":0.10*5,"Sendvic":0,"juha":0,"Pasta":0,"Cevapi":0,"Pizza":0 },
    "vodka":            {"Bloody Mary":0.045 * 5,"Gin & Tonic":0,"Martini":0,"Bloody gin":0,"Sendvic":0,"juha":0,"Pasta":0,"Cevapi":0,"Pizza":0},
    "gin":              {"Bloody Mary":0,"Gin & Tonic":0.06 * 5,"Martini":0.05 * 5,"Bloody gin":0.06 * 5,"Sendvic":0,"juha":0,"Pasta":0,"Cevapi":0,"Pizza":0},
    "tonik":            {"Bloody Mary":0,"Gin & Tonic":0.12 * 5,"Martini":0,"Bloody gin":0,"Sendvic":0,"juha":0,"Pasta":0,"Cevapi":0,"Pizza":0},
    "vermouth":         {"Bloody Mary":0,"Gin & Tonic":0,"Martini":0.01 * 5,"Bloody gin":0,"Sendvic":0,"juha":0,"Pasta":0,"Cevapi":0,"Pizza":0},
    "kruh":             {"Bloody Mary":0,"Gin & Tonic":0,"Martini":0,"Bloody gin":0,"Sendvic":0.4,"juha":0,"Pasta":0,"Cevapi":0.4,"Pizza":0},
    "sir":              {"Bloody Mary":0,"Gin & Tonic":0,"Martini":0,"Bloody gin":0,"Sendvic":0.1,"juha":0,"Pasta":0,"Cevapi":0,"Pizza":0.4},
    "mortadela":        {"Bloody Mary":0,"Gin & Tonic":0,"Martini":0,"Bloody gin":0,"Sendvic":0.1,"juha":0,"Pasta":0,"Cevapi":0,"Pizza":0},
    "vrecica juhe":     {"Bloody Mary":0,"Gin & Tonic":0,"Martini":0,"Bloody gin":0,"Sendvic":0,"juha":0.12,"Pasta":0,"Cevapi":0,"Pizza":0},
    "pasta":            {"Bloody Mary":0,"Gin & Tonic":0,"Martini":0,"Bloody gin":0,"Sendvic":0,"juha":0,"Pasta":1,"Cevapi":0,"Pizza":0},
    "maslac":           {"Bloody Mary":0,"Gin & Tonic":0,"Martini":0,"Bloody gin":0,"Sendvic":0.05,"juha":0,"Pasta":0.1,"Cevapi":0.3,"Pizza":0},
    "rajcica konzerva": {"Bloody Mary":0,"Gin & Tonic":0,"Martini":0,"Bloody gin":0,"Sendvic":0,"juha":0,"Pasta":1,"Cevapi":0,"Pizza":1},
    "cevapi":           {"Bloody Mary":0,"Gin & Tonic":0,"Martini":0,"Bloody gin":0,"Sendvic":0,"juha":0,"Pasta":0,"Cevapi":0.5,"Pizza":0},
    "brasno":           {"Bloody Mary":0,"Gin & Tonic":0,"Martini":0,"Bloody gin":0,"Sendvic":0,"juha":0,"Pasta":0.005,"Cevapi":0,"Pizza":0.8},
}

# kolicina rada u minutama
kol_rada = {"Bloody Mary":5,"Gin & Tonic":4,"Martini":6,"Bloody gin":4,"Sendvic":7,"juha":15,"Pasta":35,"Cevapi":20,"Pizza":30}


stavke_kategorije = {
    "pice":['Bloody Mary', 'Gin & Tonic', 'Martini',"Bloody gin",],
    "hrana":['Sendvic', 'juha', 'Pasta', "Cevapi", "Pizza",]
}

stavke = tuple((stavke_kategorije['pice']+stavke_kategorije['hrana']))


# cijena_sastojaka_po_potrebnoj_kolicini = kolicina sastojka potrebna * cijena sastojka 
# koliko kosta proizvodnja jedne stavke
cijena_sastojaka_po_potrebnoj_kolicini = {}

for stavka in stavke:
    cijena_sastojaka_po_potrebnoj_kolicini[stavka] = 0

for stavka in stavke:
    for sastojak in cijena_sastojaka_:
        cijena_sastojaka_po_potrebnoj_kolicini[stavka] += kol_sastojaka[sastojak][stavka] * cijena_sastojaka_[sastojak]


######################################################################################

# varijable odluke
Stvk = var('x',stavke, kind=float) # ako je type int onda traje 10 min racunanje

########################################################
#### Pomocne Varijable koje koriste varijable odluke ###
########################################################

# cijena rada = kolicina Rada * cijena minute * kolicina stavke
cijena_rada = {}
for x in stavke:
    cijena_rada[x] = kol_rada[x] * Stvk[x] * cijena_minute


# cijena sastojaka = kolicina stavke * cijena po 1 kolicini
cijena_sastojaka = {}
for x in stavke:
    cijena_sastojaka[x] = Stvk[x] * cijena_sastojaka_po_potrebnoj_kolicini[x]

# ukupna cijena stavke = cijena sastojaka + cijena rada
ukupna_cijena_stavke = {}
for x in stavke:
    ukupna_cijena_stavke[x] = cijena_sastojaka[x] + cijena_rada[x]

    
# kolicina stavke = mnozitelj porcija * stavka
# ovo je stvarna kolicina stavki 
kolicina_stavke = {}
for x in stavke:
    kolicina_stavke[x] = broj_kom_mnozitelj * Stvk[x]

ukupna_kolicina_pica =  sum(kolicina_stavke[x] for x in stavke_kategorije['pice'] )
ukupna_kolicina_hrane = sum(kolicina_stavke[x] for x in stavke_kategorije['hrana'] )

    
#  funkcija cilja :
#  provizija * ukupna_cijena_stavke - ukupna_cijena_stavke
#  maximiziramo proviziju ?
funk_cilja = {}
for x in stavke:
    funk_cilja[x] =  provizija * ukupna_cijena_stavke[x] - ukupna_cijena_stavke[x]

ukupna_cijena_sa_provizijom = {}
for x in stavke:
    ukupna_cijena_sa_provizijom[x] =  provizija * ukupna_cijena_stavke[x]
##########################


maximize( sum(funk_cilja.values()) )


# grupe uvjeta
con_kolicina_stavke = group('kol_stv') 
con_max_udio_kolicine = group('max_udio_kol')
con_min_udio_kolicine = group('min_udio_kol')
con_ukupna_cijena_stavke = group('ukupna_cijena_stavke')

# kolicina hrane/pica svakog po 100
    # donja granica
con_kolicina_stavke['Gpica']  = ukupna_kolicina_pica   >= broj_uzvanika * min_kolicina_pica_po_uzvaniku
con_kolicina_stavke['Ghrana'] = ukupna_kolicina_hrane  >= broj_uzvanika * min_kolicina_hrane_po_uzvaniku
    # gornja granica  (donja + 5%)
con_kolicina_stavke['Dpica']  = ukupna_kolicina_pica   <= broj_uzvanika * min_kolicina_pica_po_uzvaniku  * 1.05
con_kolicina_stavke['Dhrana'] = ukupna_kolicina_hrane  <= broj_uzvanika * min_kolicina_hrane_po_uzvaniku * 1.05

# max udio kolicine hrane 
for x in stavke_kategorije['hrana']:
    con_max_udio_kolicine[x] = kolicina_stavke[x] <= ukupna_kolicina_hrane * max_udio_kolicine_hrane

# max udio kolicine pica 
for x in stavke_kategorije['pice']:
    con_max_udio_kolicine[x] = kolicina_stavke[x] <= ukupna_kolicina_pica * max_udio_kolicine_pica

# min udio kolicine hrane 
for x in stavke_kategorije['hrana']:
    con_min_udio_kolicine[x] = kolicina_stavke[x] >= ukupna_kolicina_hrane * min_udio_kolicine_hrane

# min udio kolicine pica 
for x in stavke_kategorije['pice']:
    con_min_udio_kolicine[x] = kolicina_stavke[x] >= ukupna_kolicina_pica * min_udio_kolicine_pica
    
# Ukupna cijena 
con_ukupna_cijena_stavke['sve'] = sum(ukupna_cijena_sa_provizijom.values()) <= max_trosak
    

solve() 

print()
print(" Profit : %f"%vobj())
print(" Ukupna Cijena : %f"%con_ukupna_cijena_stavke['sve'].primal)



##################################################
########################## Grafovi
##################################################


plt_kolicina_stavke = {}
for x in stavke:
    plt_kolicina_stavke[x] = round(Stvk[x].primal * broj_kom_mnozitelj)

keys = np.array(list(plt_kolicina_stavke.keys()))
values = np.array(list(plt_kolicina_stavke.values()))

fig, ax = plt.subplots()    
width = 0.75 
ax.barh(keys, values, width, )
for i, v in enumerate(values):
    ax.text(v + 10, i , str(v), fontweight='bold')
plt.title('Kolicina Stavki')
plt.show()

pr_uk_hrn = 0
for x in stavke_kategorije['hrana']:
    pr_uk_hrn += plt_kolicina_stavke[x]
    
pr_uk_pc = 0
for x in stavke_kategorije['pice']:
    pr_uk_pc += plt_kolicina_stavke[x]

print(" Kolicina Hrane : %f"%pr_uk_hrn)
print(" Kolicina Pica  : %f"%pr_uk_pc)


# kol sas
plt_kolicina_sastojaka = {}

for x in cijena_sastojaka_:
    plt_kolicina_sastojaka[x] = 0
    
for stavka in stavke:
    for sastojak in cijena_sastojaka_:
        plt_kolicina_sastojaka[sastojak] += kol_sastojaka[sastojak][stavka] * Stvk[stavka].primal

for x in cijena_sastojaka_:
    plt_kolicina_sastojaka[x] = round(plt_kolicina_sastojaka[x])
    if(plt_kolicina_sastojaka[x] == 0):
        plt_kolicina_sastojaka.pop(x, None)
        
keys = np.array(list(plt_kolicina_sastojaka.keys()))
values = np.array(list(plt_kolicina_sastojaka.values()))

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
ax.pie(values, labels = keys,autopct='%1.2f%%')
print()
print(" Kolicina sastojaka : ")
plt.show()

#######################

print("radi se o porcijama pa treba ici Activity * 5 za stvarnu kolicinu")
sensitivity()
end()

## Print Uvjeta :
print()
print("Uvjeti :")
print()

print("Gornje i donje granice kolicine pica i hrane :")
print(con_kolicina_stavke['Ghrana'])
print(con_kolicina_stavke['Dhrana'])
print(con_kolicina_stavke['Gpica'])
print(con_kolicina_stavke['Dpica'])

print()
print("Ukupna Cijena :")
print(con_ukupna_cijena_stavke['sve'])

print()
print("Kolicina pojedine stavke :")
print("komentirano")

# # udio kolicine hrane 
# for x in stavke_kategorije['hrana']:
#     print(x+" : "+str(con_max_udio_kolicine[x]))

# print()
# # udio kolicine pica 
# for x in stavke_kategorije['pice']:
#     print(x+" : "+str(con_max_udio_kolicine[x]))


```
